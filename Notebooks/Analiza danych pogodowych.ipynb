# Aplikacja do monitorowania pogody w Tatrach

**Opis projektu:** Celem projektu jest stworzenie aplikacji do monitorowania aktualnych warunków dotyczących pogody i jakości powietrza w Tatrach. Powinno to obejmować takie parametry jak:
- temperatura
- temperatura odczuwalna
- wilgotność względna
- ciśnienie atmosferyczne
- zanieczyszczenie pyłami PM10

Aplikacja powinna pobierać aktualne wartości powyższych parametrów z API takich jak OpenWeatherMap czy Airly. Raz pobrane wartości powinny być zapisywane do bazy danych (tabela `weather_history`) tak aby w przyszłości można było do nich wrócić. Należy pobierać dane co godzinę i zapisywać wszystko do tabeli.

Parametry pogodowe należy zbierać dla siatki punktów rozłożonych co ok. 1km. Dane na temat zanieczyszczenia powietrza należy zbierać dla wszystkich stacji.

Aplikacja powinna posiadać podstrony:
- Aktualna pogoda
- Prognoza pogody

Prognoza pogody powinna być pobierana niezależnie od danych o warunkach rzeczywistych. Powinna być ona zapisywana do plików `.json` w osobnym folderze. Do map pogodowych użyj biblioteki folium.

Niezależnie od powyższych aplikacji przeprowadź analizę eksploracyjną historycznych wartości pogody w Tatrach.

## Pobranie i analiza danych historycznych

Pobierz dane historyczne dla tych punktów w Tatrach, dla których te dane są dostępne. Wykorzystaj do tego dowolne API, które posiada endpoint zwracający dane historyczne, np.:
- Meteostat
- Open-Meteo

1. Pobierz dane przez API. Zbierz jak najwięcej informacji.
2. Zamień na dataframe
3. Zapisz dataframe bezpośrednio do bazy danych
4. W osobnym notebooku wczytaj dane
5. Wykonaj analizę eksploracyjną

Parametry pogodowe to w więkoszosci cechy numeryczne, ale takie informacje jak nazwa stacji, dzień tygodnia albo pora roku możesz potraktować jako zmienne kategoryczne.

Analiza eksploracyjna powinna obejmować spis treści w postaci nagłówków markdown a także opis wniosków wyciągniętych a analizy. Powinna być ona zrozumiała również dla osób nietechnicznych a więc wykresy powinny być dobrze opisane.

## Baza danych do zbierania aktualnych danych pogodowych

Stwórz bazę danych o nazwie `weather` a w niej tabelę `weather_history`. Napisz zapytanie `CREATE` a następnie wykonaj go za pomocą `psycopg`. Napisz też klasę, która obsłuży dodawanie i odczytywanie rekordów z tej tabeli (każda operacja to osobna metoda).

Pobierz dane poprzez API takie jak `OpenWeatherMap` czy `Airly` i zapisz je do bazy korzystając z napisanego wcześniej konektora.

Korzystając z programu Harmonogram zadań w Windowsie zautomatyzuj wykonywanie skryptu pobierającego dane. Docelowo powinien wykonywać się on codziennie.

## Pobieranie prognozy pogody

Napisz kod, który będzie pobierał prognozę pogody z `OpenWeatherMap` dla siatki punktów i zapisywał go do plików `.json`. Zaproponuj system, w którym te prognozy będą w logiczny sposób zapisywane z plikach/folderach.

Zautomatyzuj pobieranie danych podobnie jak w poprzednim zadaniu.

## Rekomendacja AI dotycząca warunków pogodowych

Przygotuj rekomendację od AI dotyczącą dzisiejszych warunków w kontekście wędrówek górskich. W tym celu wyślij do modelu poprzez API informacje o prognozie pogody na najbliższe 24h i poproś o odpowiedź w formie JSONa (structured output) z dwoma polami:
- recommendation: safe / risky / dangerous
- justification: krótki uzasadnienie tej decyzji

## Dashboard od wizualizacji

Za pomocą agenta AI i bibliotek `streamlit` oraz `folium` stwórz dashboard, w którym znajdą się dwie strony:
- Dane historyczne
- Prognoza pogody

Zwizualizuj wybrane parametry pogodowe / zanieczyszczenia powietrza za pomocą heatmapy. 

Zintegruj dashboard z rekomendacją AI dotyczącą warunków pogodowych tak aby wyświetlić stosowny komunikat na zielonym, żółtym lub czerwonym polu (streamlit ma do tego odpowiednie komponenty).
