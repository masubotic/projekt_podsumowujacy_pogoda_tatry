# SQLite: import i odczyt danych

Notebook korzysta z funkcji `import_weather_history_csv` z `Scripts/weather_db.py`.
Obsluguje aktualny CSV z naglowkami oraz starszy format bez naglowkow.

```python
from pathlib import Path
import sqlite3
import pandas as pd

from Scripts.weather_db import initialize_weather_database, import_weather_history_csv
```

```python
db_path = Path('../data/weather.db').resolve()
csv_path = Path('../data/weather_history.csv').resolve()

initialize_weather_database(db_path)
inserted_rows = import_weather_history_csv(csv_path=csv_path, db_path=db_path, replace_existing=True)

print('DB:', db_path)
print('CSV:', csv_path)
print('Zaimportowano wierszy:', inserted_rows)
```

```python
query = '''
SELECT
    id,
    observed_at,
    location,
    temperature_c,
    humidity,
    pressure_hpa,
    wind_speed_ms,
    weather_description,
    created_at
FROM weather_history
ORDER BY observed_at DESC
'''

with sqlite3.connect(db_path) as conn:
    df = pd.read_sql_query(query, conn)

print('Liczba rekordow w tabeli:', len(df))
df.head()
```
